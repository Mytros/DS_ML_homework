### Домашні завдання на відпрацювання поняття функції та її похідної

1. Задана функція залежності оцінки від кількості годин, витрачених на навчання: $ f(x) = 10x - 5 $.
  1. З допомогою Python обчисліть похідну цієї функції.
  2. Використовуйте бібліотеку `matplotlib` для побудови графіків функції та її похідної в діапазоні значень від 0 до 10 годин. Проаналізуйте, як зміна кількості годин вивчення впливає на оцінку.



```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

# Визначаємо символічну змінну
x = sp.symbols('x')

# Задана функція
f = 10 * x - 5

# Обчислення похідної
derivative_f = sp.diff(f, x)

# Виводимо функцію та похідну
print("Функція f(x) = 10x - 5")
print("Похідна f'(x):", derivative_f)

# Значення x від 0 до 10
x_vals = np.linspace(0, 10, 10)

# Обчислюємо значення функції вручну
y_vals = 10 * x_vals - 5

# Похідна — константа 10
y_prime_vals = np.full_like(x_vals, 10)

# Побудова графіків
plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, label='f(x) = 10x - 5', color='blue')
plt.plot(x_vals, y_prime_vals, label="f'(x) = 10", linestyle='--', color='red')
plt.title('Графік функції та її похідної')
plt.xlabel('Кількість годин навчання (x)')
plt.ylabel('Оцінка')
plt.legend()
plt.grid(True)
plt.show()
```

Висновок:

Зміна кількості годин вивчення лінійно впливає на оцінку. Чим більше годин витрачено на навчання, тим вища оцінка. Кожна 1 година навчання дає приріст до оцінки 10 балів.

2. Використовуючи Python і `SymPy`, обчисліть похідні для функцій $ f(x) = \sin(x) $ та $ f(x) = \ln(x) $. Зобразіть ці функції та їхні похідні за допомогою `matplotlib`.



```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

# Символьна змінна
x = sp.symbols('x')

# Задані функції
f_sin = sp.sin(x)
f_ln = sp.ln(x)

# Похідні функцій
derivative_sin = sp.diff(f_sin, x)
derivative_ln = sp.diff(f_ln, x)

# Вивід у консоль
print("Похідна f(x) = sin(x):", derivative_sin)
print("Похідна f(x) = ln(x):", derivative_ln)

# Значення x для побудови графіків
x_vals = np.linspace(0.1, 10, 400)  # уникати 0 для ln(x)
x_vals_trig = np.linspace(-2 * np.pi, 2 * np.pi, 400)

# Значення функцій
y_sin = np.sin(x_vals_trig)
y_sin_prime = np.cos(x_vals_trig)

y_ln = np.log(x_vals)
y_ln_prime = 1 / x_vals

# Побудова графіків
plt.figure(figsize=(14, 6))

# Графік sin(x) та його похідної
plt.subplot(1, 2, 1)
plt.plot(x_vals_trig, y_sin, label='f(x) = sin(x)', color='blue')
plt.plot(x_vals_trig, y_sin_prime, label="f'(x) = cos(x)", linestyle='--', color='red')
plt.title('Функція sin(x) та її похідна')
plt.xlabel('x')
plt.ylabel('Значення')
plt.legend()
plt.grid(True)

# Графік ln(x) та його похідної
plt.subplot(1, 2, 2)
plt.plot(x_vals, y_ln, label='f(x) = ln(x)', color='blue')
plt.plot(x_vals, y_ln_prime, label="f'(x) = 1/x", linestyle='--', color='red')
plt.title('Функція ln(x) та її похідна')
plt.xlabel('x')
plt.ylabel('Значення')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```


3. Загальні витрати компанії від обсягу виробництва задані функцією $ f(x) = x^2 - 5x + 1 $, де $ x $ — кількість вироблених одиниць. Обчисліть похідну цієї функції з допомогою SymPy та візуалізуйте значення функції і її похідної на відрізку від 0 до 100.

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

# Визначаємо символічну змінну
x = sp.symbols('x')

# Функція загальних витрат
f = x**2 - 5*x + 1

# Обчислення похідної
derivative_f = sp.diff(f, x)

# Вивід у консоль
print("Функція загальних витрат f(x) = x² - 5x + 1")
print("Похідна f'(x):", derivative_f)

# Побудова графіків
x_vals = np.linspace(0, 100, 100)
y_vals = x_vals**2 - 5*x_vals + 1
y_prime_vals = 2*x_vals - 5  # похідна: f'(x) = 2x - 5

# Графік
plt.figure(figsize=(12, 6))
plt.plot(x_vals, y_vals, label='f(x) = x² - 5x + 1', color='blue')
plt.plot(x_vals, y_prime_vals, label="f'(x) = 2x - 5", linestyle='--', color='red')
plt.title('Функція загальних витрат і її похідна')
plt.xlabel('Кількість вироблених одиниць (x)')
plt.ylabel('Витрати')
plt.legend()
plt.grid(True)
plt.show()
```

Функція f(x) = x² − 5x + 1 — квадратична, тобто витрати зростають із ростом обсягу виробництва, але спочатку мають спад.

Похідна f'(x) = 2x − 5:

Якщо x < 2.5, похідна від’ємна → витрати зменшуються.

Якщо x > 2.5, похідна додатна → витрати починають зростати швидше.

```python
# variant 2
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

# Символьна змінна
x = sp.symbols('x')

# Функція та похідна
f = x**2 - 5*x + 1
f_prime = sp.diff(f, x)

# Точка, де похідна = 0
a = 2.5
f_a = a**2 - 5*a + 1  # значення функції в точці a
slope = 2*a - 5       # значення похідної в точці a (0 у цьому випадку)

# Рівняння дотичної: y = f(a) + f'(a)*(x - a)
x_vals = np.linspace(0, 10, 500)
y_vals = x_vals**2 - 5*x_vals + 1
y_prime_vals = 2*x_vals - 5
tangent_vals = f_a + slope * (x_vals - a)  # пряма дотична

# Побудова графіків
plt.figure(figsize=(12, 6))
plt.plot(x_vals, y_vals, label='f(x) = x² - 5x + 1', color='blue')
plt.plot(x_vals, y_prime_vals, label="f'(x) = 2x - 5", linestyle='--', color='red')
plt.plot(x_vals, tangent_vals, label='Дотична в x = 2.5', linestyle=':', color='green')
plt.axvline(x=a, color='gray', linestyle=':', label='x = 2.5 (мінімум)')
plt.scatter([a], [f_a], color='black', zorder=5)  # точка дотику

plt.title('Функція витрат, похідна та дотична в x = 2.5')
plt.xlabel('Кількість вироблених одиниць (x)')
plt.ylabel('Витрати')
plt.legend()
plt.grid(True)
plt.show()
```
